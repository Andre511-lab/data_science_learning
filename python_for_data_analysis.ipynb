### Creating ndarray

The easiest way to create an array is to use the array function. This accept any sequence-like object (including other arrays) and produces a new NumPy array containing the passed data. For example, a list is a good candidate for conversion :

```python
import numpy as np
```

```python
data1 = [6, 7.5, 8, 0, 1]
arr1 = np.array(data1)
arr1
```

Nested sequences, like a list of equal-length lists, will be converted into a multidimensional array :

```python
data2 = [[4, 3, 2, 1], [8, 7, 6, 5]]
arr2 = np.array(data2)
arr2
```

```python
arr2.ndim
```

```python
arr2.shape
```

Unless explicitly specified,  np.array tries to infer a good data type for the array that it creates. The data type is stored in a special dtype object; for example, in the above two examples we have :

```python
arr1.dtype
```

```python
arr2.dtype
```

In addition to np.array, there are a number of other functions for creating new arrays. As examples, zeros and ones create arrays of 0's or 1's, respectively, with a given length or shape. empty creates an array without initializing its value to any particular value. To create a higher dimensional array with these methods, pass a tuple for the shape :

```python
np.zeros(10)
```

```python
np.zeros((3, 6))
```

```python
np.empty((2, 3, 2))

# Itâ€™s not safe to assume that np.empty will return an array of all zeros. In
# many cases, as previously shown, it will return uninitialized garbage
# values.
```

```python
points = np.arange(-5, 5, 0.01)
```

```python
xs, ys = np.meshgrid(points, points)
ys
```

```python
z = np.sqrt(xs ** 2 + ys ** 2)
z
```

```python
plt.imshow(z, cmap=plt.cm.gray); plt.colorbar()
```

```python
plt.imshow(z, cmap=plt.cm.gray); plt.colorbar()
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid values")
```
